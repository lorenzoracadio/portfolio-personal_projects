```python
### CREATING THE DATABASE AND TABLE

# Create/Connect the SQLite Database
conn = sqlite3.connect('SLF.db')

# Create a cursor object to execute SQL commands
cursor = conn.cursor()

# Create the table
cursor.execute('''
    CREATE TABLE IF NOT EXISTS SLFClients (
        id INTEGER PRIMARY KEY,
        Year INTEGER,
        Month INTEGER,
        Process_Date TEXT,
        Policy_Owner TEXT,
        Policy_Number NUMBER
        Product TEXT,
        Gender TEXT,
        Age INTEGER,
        Location TEXT,
        Social_Status TEXT,
        Premium_Mode TEXT,
        Source TEXT
    )
''')

# Confirm changes
conn.commit()
conn.close()
```
